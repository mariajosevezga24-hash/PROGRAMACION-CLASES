Diccionarios

Los diccionarios guardan información en pares **clave**:**valor**. No se accede por indice sino por clave

```python
alumno = {"nombre": "Juan", "edad":18, "curso": "4° año"}
alumno
```

```python
alumno["edad"]
```

```python
alumno["nombre"]
```

```python
len(alumno)
```

```python
alumno["edad"] = 19
alumno
```

```python
alumno["nota"] = 9 #Se pueden agregar nuevas claves a diccionarios
alumno
```

```python
alumno["nacionalidad"]= "Colombiano"
alumno
```

```python
del alumno["curso"] #Tambien se pueden eliminar claves
alumno
```

```python
alumno["Años"]= alumno["edad"]
alumno
```
